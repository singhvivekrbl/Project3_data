```python
def load_images_from_folder(folder):
    images = []
    filenames = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename))
        if img is not None:
            images.append(img)
            filenames.append(filename)
    return images, filenames
```

```python
def detect_faces(input_path: str) -> dict:
    result_list = []
    '''
    Your implementation.
    '''
    # imgs is the img array 
    imgs, filenames = load_images_from_folder(input_path)
    no_of_imgs = len(imgs)
    # print(length)
    face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

    for i in range(0,no_of_imgs):
        # Load the cascade
        # Convert into grayscale
        gray = cv2.cvtColor(imgs[i], cv2.COLOR_BGR2GRAY)
        # Detect faces in the image
        face_loc = face_cascade.detectMultiScale(
            gray,
            scaleFactor=1.23,
            minNeighbors=5
            
        )

        face_locations = [x + 2 for x in face_loc]

        no_of_faces = len(face_locations)
                
        for j in range(0,no_of_faces):
            x, y, width, height = face_locations[j]    
            temp_list = [int(x), int(y), int(width), int(height)]
           
            name = filenames[i]
       
            temp_dict = {}
            
            temp_dict["iname"] = filenames[i] 
         
            temp_dict["bbox"] = temp_list
            
            result_list.append(temp_dict)
            
    return result_list

detect_faces(folder)
```

```python
from sklearn.cluster import KMeans
def cluster_faces(input_path: str, K: int) -> dict:
    result_list = []

    detect_face = detect_faces(input_path)
    data_cluster = []
    image_path = []
    for i in detect_face:
        
        img_path = input_path+ i['iname']
        
        img = cv2.imread(img_path)
        
        rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        
        a = i['bbox']
        # print( a)
        # print("yes")

        l = i['bbox'][0]
        # print(l)
        t = i['bbox'][1]
        r = l+i['bbox'][2]
        b = t+i['bbox'][3]
        box = (t,r,b,l)
        # print(box)

        feature_vector = face_recognition.face_encodings(rgb,[box])
        data_cluster.append(list(feature_vector[0]))
        # print(data_cluster)
        image_path.append(i['iname'])

    kmeans=KMeans(int(K))
    kmeans.fit(data_cluster)
    
    identify_cluster = kmeans.fit_predict(data_cluster)
    # print(identify_cluster)

    for j in range(K):
        temp_arr = []
        temp_dict = {}
        # get the name of all the images within a cluster
        for i in range(0,len(identify_cluster)):
            if(identify_cluster[i] == j):
                # temp_arr.append(img_name[i])   
                temp_arr.append(image_path[i])

        temp_dict["cluster_no"] = j
        temp_dict["elements"] = temp_arr     
        result_list.append(temp_dict)
    for h in result_list:
        hc= []
        clust  = h['cluster_no']
        for name in h['elements']:
            img= cv2.imread(input_path+ name)

            img =  cv2.resize(img,(50,50), interpolation = cv2.INTER_CUBIC)
            hc.append(img)
        h_img = cv2.vconcat(hc)
        cv2.imwrite('cluster_'+str(clust)+'.jpg',h_img)
        cv2.imshow('clustered',h_img)
        cv2.waitKey(1000)
```
